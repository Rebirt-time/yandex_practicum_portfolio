```python
def pass_value_barh(df):
    (
        (df.isna().mean()*100)
        .to_frame()
        .rename(columns = {0:'space'})
        .query('space > 0')
        .sort_values(by = 'space', ascending = True)
        .plot(kind= 'barh', figsize=(25,4), rot = -5, legend = False, fontsize = 30)
        .set_title('Пропущенные значения' + "\n", fontsize = 30, color = 'g')    
    );    
```

```python
pass_value_barh(data)
```

#### Поработаем более детально со столбцом 'user_rating'. Разберёмся, что такое 'tbd'.Прочёл из источника в интернете: https://autogear.ru/article/171/345/tbd---chto-znachit-eto-sokraschenie/ .To Be Determined, то есть "Будет определено". Очень часто подобное сокращение можно встретить в анонсах компьютерных игр в графе "Дата выхода". Когда разработчик той или иной игры уже сообщил о том, что она готовится к выходу, но пока еще не может назвать точную дату, именно это сокращение и используется. Поэтому, пока игра готовится к выходу,её ещё нет,замени 'tbd' на np.NaN. После переведём столбец из object во float, так как оценка не может быть объектом, а по условию у нас она максимум 10.

```python
# Рассмотрим столбец с оценкой пользователя,выведем уникальные значения на экран 
data['user_rating'].unique()
```

```python
# Посмотрим сколько таких значений у нас 
data[data['user_rating'] == 'tbd']['user_rating'].count()
```

```python
# определим,для специального присвоения значение NaN элементу структуры данных, используем np.NaN
data.loc[data['user_rating'] == 'tbd','user_rating'] = np.NaN
```

```python
# Оценка пользователей не может быть object , поскольку есть числа вещественные, сделаем float 
data['user_rating'] = data['user_rating'].astype('float')
```

```python
# Заполним наши NaN на -1 ,поставим, так как отрицательного рейтинга не бывает 
data['user_rating'] = data['user_rating'].fillna(-1)
```

```python
# Проверим и выведем на экран 
data['user_rating'].sort_values().unique()
```

```python
# Выведем информацию о таблице 
data.info()
```

```python
# Посмотрим на пропуски от оценки критиков 
data['critics_score'].unique()
# Оценка по условию должна быть максимум 100, проверим  
data[data['critics_score'] <= 100]
# Заполним пропуски поставим заглушку -1, так как оценки критиков могут быть разные на разных игровых платформах 
data['critics_score'] = data['critics_score'].fillna(-1).astype(int)
# Выведем на экран и проверим, предварительно проверил,что в данном столбце нет дробных, перевёл в int  
data['critics_score'].sort_values().unique()
```

#### В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков»

```python
# Выведем столбец на экран и посмотрим на значения 
data.esrb_rating.value_counts().to_frame()
```

```python
# Заполним пропуски заглушкой, но сделаем -1 ,чтобы не было  больших чисел  
data['esrb_rating'] = data['esrb_rating'].fillna('undefined')
```

```python
# Рассмотри очень интересную категорию, она покажет в какие игры рекомегдуется играть людям разного возраста 
# Составим функцю по ESRB рейтингу и разобьём на категории.  
def category_of_esrb(row):
    try:
        if 'E' in row or 'E10+' in row or 'K-A' in row:
            return 'Любой возраст'
        elif 'T' in row:
            return 'от 13 лет'
        elif 'M' in row:
            return 'от 17 лет'
        elif 'AO' in row:
            return 'от 18 лет'
        elif 'EC' in row:
            return 'от 3 лет'
    except:
        return 'Ожидает оценки '
    
# Добавим столбец категорий с рейтингом в таблицу 
data['category_of_esrb'] = data['esrb_rating'].apply(category_of_esrb)
```

```python
# Посмотрим,что получилось, выведем 10  строк на экран 
data.head()
```

## Проверим гипотезы

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые:

```python
# Выводим таблицу на экран для наглядности 
new_data_2012.head()
```

Применин специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них с актуальными годами. Сравним  независимые выборки: применим метод  scipy.stats.ttest_ind()

    Сформулируем гипотезы:
    Нулевая гипотеза: средние рейтинги платформ XOne и PC одинаковые
    Альтернативная гипотеза: средние рейтинги платформ XOne и PC разные

```python
# Проработаем пропуски и выделим платформы XOne и PC удалим NAN в пользовательских оценках  
pc_t = new_data_2012[(new_data_2012['game_platforms'] == 'PC')]
xone_t = new_data_2012[(new_data_2012['game_platforms'] == 'XOne')]
# отфильтруем данные 
pc_t = pc_t[pc_t['user_rating'] != -1]
xone_t = xone_t[xone_t['user_rating'] != -1]
xone_t = xone_t['user_rating'].dropna()
pc_t = pc_t['user_rating'].dropna()
```

```python
# Комментарий ревьюера
xone_t.min(), pc_t.min()
```

```python
# Выборки разного размера или можно предположить,
# дисперсии у генеральных совокупностей разные 
# укажим equal_var = False испольщуем вызов метода scipy.stats.ttest_ind(). 

alpha= 0.05 # критический уровень статической значимости 
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(xone_t, pc_t, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, средние рейтинги платформ XOne и PC одинаковые")
else:
    print("Не получилось отвергнуть нулевую гипотезу, средние рейтинги платформ XOne и PC разные")    
```

```python
# Посмотрим более детально, выведем оба среднего рейтинга на экран округлим значения 
print(f'Средний рейтинг: XOne {xone_t.mean():.2f}')
print(f'Средняий рейтинг: PC {pc_t.mean():.2f}')
```

Ответ: средние рейтинги платформ XOne и PC одинаковые
Средний рейтинг платформы PC равен среднему рейтингу платформы XOne

###  Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Применин специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них с актуальными годами. Сравним  независимые выборки: применим метод  scipy.stats.ttest_ind()

    Сформулируем гипотезы:
    Нулевая гипотеза: средние рейтинги жанров Action и Sports одинаковые.
    Альтернативная гипотеза: средние рейтинги жанров Action и Sports не одинаковые. 

```python
# Проработаем пропуски и выделим жанры Action и Sports удалим NAN в пользовательских оценках  
action_t = new_data_2012[(new_data_2012['game_genre'] == 'Action')]
sports_t = new_data_2012[(new_data_2012['game_genre'] == 'Sports')]
# Отфильтруем данные
action_t = action_t[action_t['user_rating'] != -1]
sports_t = sports_t[sports_t['user_rating'] != -1]
action_t = action_t['user_rating'].dropna()
sports_t = sports_t['user_rating'].dropna()
```

```python
# Выборки разного размера или можно предположить,
# дисперсии у генеральных совокупностей разные 
# укажим equal_var = False испольщуем вызов метода scipy.stats.ttest_ind(). 
alpha= 0.05 # критический уровень статической значимости 

results = st.ttest_ind(action_t, sports_t, equal_var = False)
# если p-value окажется меньше него - отвергнем гипотезу
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, средние рейтинги жанров Action и Sports одинаковые.")
else:
    print("Не получилось отвергнуть нулевую гипотезу,  средние рейтинги жанров Action и Sports не одинаковые")    
```

```python
# Посмотрим более детально, выведем оба среднего рейтинга на экран округлим значения 
print(f'Средний рейтинг: Action {action_t.mean():.2f}')
print(f'Средняий рейтинг: Sports {sports_t.mean():.2f}')
```

Ответ: средние рейтинги жанров Action и Sports не одинаковые
Средний рейтинг жанра Action выше, но не значительно, но тем не менее пользователям больше нравится Action
