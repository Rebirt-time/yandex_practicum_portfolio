Посчитаем и добавим в таблицу: тип этажа квартиры (значения — «первый», «последний», «другой»)

```python
# Напишем функцию для колонки типа этажа квартиры:
"""
-'Первый', если этаж равен 1:
  -'Последний', если этаж равен максимальному этажу в доме:
  -'Другой'

"""
def category_of_floor (row):
    
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'

# Добавим столбец category_of_floor к датафрейму    
# Применим метод apply к каждой строке датафрейма через axis=1      
data['category_of_floor'] = data.apply(category_of_floor, axis=1)
```

```python
# Выведем на экран получившуюся таблицу
display(data.head())
```

```python
# Посчитаем и добавим в таблицу расстояние в км до центра города
# Поскольку в условии есть ещё и расстояние до центра в метрах, удалять столбец не будем
# Округлим до целых чисел
data['city_centers_nearest_km'] = data['city_centers_nearest'] / 1000
data['city_centers_nearest_km'] = data['city_centers_nearest_km'].round()
```
